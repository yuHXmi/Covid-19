```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_1.csv')

df.head(10)
```

```python
plt.scatter(df["Red blood Cells"], df["Leukocytes"])

plt.title("Scatter Plot of Red blood Cells vs Leukocytes")
plt.xlabel("Red blood Cells")
plt.ylabel("Leukocytes")

# Show the plot
plt.show()
```

```python
sns.pairplot(df[['Red blood Cells', 'Leukocytes']])
```

```python
plt.scatter(df['Red blood Cells'], df['Leukocytes'], c=df['SARS-Cov-2 exam result'], cmap='viridis')

# Adding title and labels
plt.title("Scatter Plot (Color-coded by SARS-Cov-2 exam result)")
plt.xlabel("Red blood Cells")
plt.ylabel("Feature2")

# Show color bar
plt.colorbar(label='SARS-Cov-2 exam result')

# Display the plot
plt.show()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(16, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
